```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('train.csv')
test =pd.read_csv('test.csv')
df.head()
```

```python
nan_col = [i for i in df.columns if df[i].isnull().any()]
for i in nan_col:
    print(i, df[i].isnull().sum())
```

```python
dropped_cols = ['Alley', 'PoolQC', 'Fence', 'MiscFeature', 'FireplaceQu']
df = df.drop(dropped_cols, axis = 1)
test = test.drop(dropped_cols, axis = 1)
```

```python
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
test = test.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df.head()
```

```python
df = pd.get_dummies(df).astype(int)
test = pd.get_dummies(test).astype(int)
```

```python
com_col = (df.columns).intersection(test.columns)
dif_col = (df.columns).difference(test.columns)
```

```python
test.head()
```

```python
print(dif_col)
```

```python
for i in dif_col:
    test.insert(251,i,0)
test = test.drop('SalePrice',axis=1)
```

```python
test = test.reindex(sorted(test.columns),axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=400)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('SalePrice', axis = 1)
X = X.reindex(sorted(X.columns),axis=1)
y = df['SalePrice']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, y_pred))
```

```python
model.fit(X, y)
```

```python
result = model.predict(test)
```

```python
test['SalePrice'] = result
submission = test[['Id','SalePrice']]
submission.to_csv('submission.csv',index=False)
```
